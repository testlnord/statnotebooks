```python
%%R
library(MASS)
library(lattice)
library(latticeExtra)
library(e1071)
library(nnet)
```

Загрузка данных

```python
%%R
seeds.data  <- read.table("seeds_dataset.txt")
seeds.data$V8 <- factor(seeds.data$V8, levels=c(2,1,3))
print(head(seeds.data))
```

Попробуем метод "в лоб"

```python
%%R
lm1 <- tune(lda, V8 ~ ., data = seeds.data, 
            predict.func = function(...) predict(...)$class, 
            tunecontrol = tune.control(sampling = "fix", fix = 2/3))
print(lm1)
```

```python
%%R
print(marginal.plot(~ seeds.data, data= seeds.data, group=V8))
print(splom(~seeds.data, data = seeds.data, groups = V8))
```

Они отлично делятся по сортам, если делить их по сортам :) 

Еще хорошо заметно, что некоторые из измерений сильно зависимы, причем линейно. Давайте их выкинем.

```python
%%R
print(cor(seeds.data[-8]))
```

Некоторые из корреляций близки к единице. Все предикторы делятся на группы по "похожести". 
Оставим из каждой группы только одного представителя.

```python
%%R
lm2 <- tune(lda, V8 ~ V1 + V3 + V6 + V7, data = seeds.data, 
            predict.func = function(...) predict(...)$class, 
            tunecontrol = tune.control(sampling = "fix", fix = 2/3))
print(lm2)
```

Стало лучше. Давайте так и оставим.

Попробуем теперь побаловаться с мультиномиальной регрессией

```python
%%R
mln1 <- multinom(V8 ~ ., data = seeds.data, trace = FALSE)
print(summary(mln1))
print(tune(multinom, V8 ~ ., data = seeds.data, trace = FALSE,))
```

```python
%%R
mln2 <- multinom(V8 ~  V1 + V3 + V6 + V7, data = seeds.data, trace = FALSE)
print(summary(mln2))
print(tune(multinom, V8 ~ V1 + V3 + V6 + V7, data = seeds.data, trace = FALSE,))
```

Внезапно стало хуже. Видимо я зря что-то выкинул. А давайте спросим у Акаике, что нам следовало бы выкинуть.

```python
%%R
print(stepAIC(mln1))
```

```python
%%R
print(tune(multinom, V8 ~ V1 + V4 + V6 + V7, data = seeds.data, trace = FALSE,))
```

Вот так дела! Мы заменили один признак на другой, причем на коррелирующий с уже представленными и все заиграло. 

```python
%%R
print(tune(multinom, V8 ~ V1 + V6 + V7, data = seeds.data, trace = FALSE,))
print(tune(lda, V8 ~ V1 + V6 + V7, data = seeds.data, 
            predict.func = function(...) predict(...)$class, 
            tunecontrol = tune.control(sampling = "fix", fix = 2/3)))
```

По сути все полученные модели одинаковы по качеству. Кросвалидация показывает на них схожие значения ошибки, которые от запуска к запуску варьируются в одном диапазоне.
