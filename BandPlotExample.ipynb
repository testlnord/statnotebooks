```python
%%R
ex.data = data.frame(d = rnorm(100, 10, 100), x = 1:100)
ex.data$h = ex.data$d + ex.data$x
ex.data$l = ex.data$d - ex.data$x
print(bandplot(l+h ~ x, data = ex.data))
```

```python
%%R
print(bandplot(l+h~ x |d > 10, data = ex.data))
```
