```python
%%R
library(e1071)
library(MASS)
library(lattice)
library(latticeExtra)
```

```python
%%R
gl.data <- read.csv2("GlaucomaMVF.txt", sep=" ", header=T,comment.char = "#", )
print(head(gl.data$Class))
```

Предикторов очень много, ничего рисовать не хочется, думать тоже. Сразу делаю tune и svm.

*tune делал, подбирал параметры, в голом R, в браузере это работает вечность.*

```python
%%R
svm.gaussian.log <- tune.svm(Class ~ ., data = gl.data, kernel="radial", cost = 4,  gamma = 0.00390625)
```

```python
%%R
svm.linear.log <- tune.svm(Class ~ ., data = gl.data, kernel="linear")
```

```python
%%R
svm.polynomial.log <- tune.svm(Class ~ ., data = gl.data, kernel="polynomial", cost = 0.00390625, gamma = 0.125)
```

```python
%%R
print(svm.gaussian.log)
print(svm.linear.log)
print(svm.polynomial.log)
```

На представленных данных радиальное и линейное ядра отработали почти одинаково. А вот полиномиальное сильно хуже.
